```python
text=text.lower()
```

```python
nltk.download('stopwords')
```

```python
punct = string.punctuation
```

```python
text =text.split('.')
# to get the data in sentences, 
# this is not an ideal approach as if we have ph.d then it would fail, so we need to use
# nltk.sent_tokenize function to split our data 
```

```python
translator=str.maketrans('','',punct)
text1=[]
for i in text:
    clean_text = i.translate(translator)
    text1.append(clean_text)
```

```python
text1
```

```python
stopwords=set(nltk.corpus.stopwords.words('english'))
```

```python
temp_text=[]
for i in text1:
    a=i.split()
    l=[]
    for b in a:
        if b not in stopwords and punct:
            l.append(b)
    temp_text.append(' '.join(l))
    
```

```python
tokenizer= Tokenizer()
```

```python
final_text_withscores=[]
for i in temp_text:
    l=[]
    m=0
    for j in i.split():
        m=m+ final_df[final_df['Words']==j]['Weights'].values[0]
        l.append(j)
    final_text_withscores.append(' '.join(l)+" "+str(m)) 
    # here we are able to get the scores of each sentence 
```

```python
final_text_withscores
# the only problem That I am fcaing in this would be, if by chance a complete line is removed then the weights will be assigned 
# wrongly to, overcome that problem we need at the time of assignmnet when we split the data based of . I need to assign the data 
# each line a number even if the line gets removed the number is not.
```

```python
scores=[]
for i in range(len(final_text_withscores)):
    scores.append(float(final_text_withscores[i].split()[-1]))
```

```python
scores1=[]
for i in range(len(scores)):
    scores1.append(scores[i]/sum(scores))
    

# these are scores based of each line that we have, In this we can keep all the lines that are above 3 
```
