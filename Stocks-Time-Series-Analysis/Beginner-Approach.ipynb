# Arima Model [Autoregressive intigrated moving avarerage]

```python
# It uses the time series past values and/or forecast error to predict it's future value 
# Arima model assumption is to have the stationery time series 
# AR+I+MA
# p+d+q
# Arima is broken down into 3 parameters that are p,d and q
# we have to install a module name pmdarima from cmd or anaconda prompt 
# from pmdarima import auto_arima
# arima_fit=auto_arima(data['Price'],start_p=0,start_q=0, max_p=8,max_q=8,trace=True)
# here p represents autoregressive[AR] term[use of past values of the series in prediction equation] and 
# d[I] represents degree of differencing[differencing is the process of transforming a non stationery time series into stationery 
# by taking the differences between consecutive observations]
# If not stationery, the time series can be differenced to become stationery 
# q[MA] represents moving averages[use of past forecast errors in the prediction equation] 
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(series):
    result=adfuller(series,autolag='AIC')
    labels =['Test stats','P-value','#lags used','# observations']
    out=pd.Series(result[0:4],index=labels)
    print(out.to_string()) 
    if result[1] <= 0.05:
        print('reject the null hypothesis,i.e data has no unit root and is stationery')
    else:
        print('failed to reject null hypothesis, data has a unit root and is non stationery')
adf_test(new_data['price'])        
```
