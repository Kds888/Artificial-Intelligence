STOCHASTIC GRADIENT DESCENT 
STOCHASTIC GRADIENT DESCENT COMES TO PLAY WHEN WE HAVE A LARGER DATASET WITH FEATURES LETS SAY 23000 AND ENTERIES TO 1 MILLION IN THIS CASE GRADIENT DESCENT WILL HAVE TO PERFORM CALCLUATIONS FOR EVERY FEATURE AND WITH RESPECT TO EVERY DATA POINT, WHICH IS A LOT FOR GRADIENT DESCENT SO HERE WE GO FOR STOCHASTIC GRADIENT DESCENT. 
AND STOCHASTIC GRADIENT WILL JUST REDUCE THE NUMBER OF STEPS IN CALCULATIONS BEACUSE IT WILL RANDOMLY PICK A ROW OF VALUES FROM THE DATASET AND INSTAED OF CACULATING BY TAKING IN FOR EVERY VALUE IN THE DATASET IT WILL JUST CHOOSE A RANDOM VALUE.
IT IS VERY USEFUL IF WE HAVE REDUNDENCIES IN DATA BECAUSE WE CAN CHOOSE ONLY OF THE DATA POINTS FROM THE CLUSTERS AND THEN CAN CALCULATE THE VALUES BASED OF THE LOSS FUNCTIONS.


WE CAN ALSO TAKE A MINI BATCH TO ESTIMATE THE PARAMETERS AND USING A MINI BATCH CAN RESULT IN MORE STABLE ESTIMATES OF THE PARAMETERS IN FEWER STEPS. 
STOCHASTIC GRADIENT HELP ADJUST TO THE NEW DATA EASILY.BASICALLY WE CAN ADD NEW DATA POINTS AND IT WILL ADJUST THE NEW LINE ACCORDINGLY.
STOCHASTIC GRADIENT DESCENT IS BASICALLY SIMILAR TO GRADIENT DESCENT, EXCEPT IT ONLY LOOKS AT ONE SAMPLE PER STEP, OR A SMALL SUBSET, MIN-BATCH FOR EACH STEP.
